```python
import pandas as pd

speeds_df = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/visualisation/speeds_data.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
unique_rds = speeds_df['road'].unique()
unique_rds
```

```python
for road in unique_rds:
  subset = speeds_df[speeds_df['road'] == road]
  subset.plot(kind='bar', x='road', y='speed')
  plt.ylabel('Speed')
  plt.title(f'Speed for Road {road}')
  plt.show()
```

```python
mean_speeds_df = speeds_df.groupby(by='road')['speed'].mean()
mean_speeds_df
```
